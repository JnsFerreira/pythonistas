```python
def square(n: int) -> int:
  """
  Blbla bla
  """
  return n**2

print(square(2))


s = square
print(s(2))

x = 20
a = 3

def func1(n):
  print(n*1)

def func2(n):
  print(n*2)

def func3(n):
  print(n*3)

if a == 0:
  func1(x)

elif a == 3:
  func2(x)

elif a == 5:
  func3(x)

functions_translation = {
    0: func1,
    3: func2,
    5: func3,
}

functions_translation.get(a)(x)

def fatorial(n):
  if n == 1:
    return 1
  
  return n * fatorial(n-1)

print(fatorial(5))


def raise_example(num_1):
  if num_1 < 0:
    raise ValueError('num is below 0')
  
  return num_1

def exception_treatment(exc):
  print(exc)


try:
  print(raise_example(-1))
except Exception as exc:
  exception_treatment(exc)
finally:
  print('acabei')


def generator_ex():
  for i in range(10):
    yield i

sum = 0
for i in generator_ex():
  sum += i

print(sum)

# decorators.


```
