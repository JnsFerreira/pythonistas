```python
"""
Polynomial

Ex.: 3x^2 + 4x + 7
     x^2 - 3x + 4
"""
class Polynomial:
    
    def __init__(self, *coeffs):
        self.coeffs = coeffs
    
    def __repr__(self):
        return f'Polynomial(*{self.coeffs})'

    def __add__(self, other):
        return Polynomial(*[x + y for x, y in zip(self.coeffs, other.coeffs)])

    def calc(self, x):
        return sum([
            (self.coeffs[0] * x) ** 2,
            self.coeffs[1] * x,
            self.coeffs[2]
        ])    
```

```python
p1 = Polynomial(3, 4, 7)

print(p1)
```

```python
print(p1.coeffs)
```

```python
p2 = Polynomial(1, -3, 4)
p2.calc(x=2)
```

```python
(p1 + p2).calc(x=2)
```

```python
"""
Leitor de arquivos
"""
class Leitor:

    breakpoint = 0

    def ler(self, file_path, size=None):
        file = self.abrir_arquivo(file_path)
        
        for row in list(file)[self.breakpoint:self.breakpoint + size]:
            self.breakpoint += 1
            print(row, end="")

    def abrir_arquivo(self, file_path):
        return open(file_path, 'r')
```

```python
l1 = Leitor()

l1.ler(
    file_path='/content/sample_data/california_housing_test.csv', 
    size=10
)
```

```python
l1.ler(
    file_path='/content/sample_data/california_housing_test.csv', 
    size=10
)
```

```python
l2 = Leitor()
l2.ler('/content/sample_data/mnist_test.csv', size=10)
```
