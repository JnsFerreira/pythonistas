Loops in Python

while

```python
index = 0
word = 'abacate'
letter = word[index]
while letter != 't':
  print(letter)
  index += 1
  letter = word[index]
```

for

```python
word = 'abacate' # tuples, lists, sets, any iterable
for letter in word:
  print(letter)
```

```python
matrix = [ [0, 1], [2, 3] ]  # O(n2)
for line in matrix:
  print(line)
  for colum in line:
    print(colum)
```

```python
ingredients_list = [
  {
      'name': 'apple',
      'quantity': 1
  },
  {
      'name': 'banana',
      'quantity': 0
  },
  {
      'name': 'pineapple',
      'quantity': 1
  },
]
for ingredient in ingredients_list:
  if not ingredient.get('quantity'):  # same as == 0, False, [], {}
    if ingredient.get('name') == 'banana':
      continue  # get next element

    print(f'problem, no {ingredient.get("name")}')
    break  # stop loop
print('end')
```

```python
ingredients_list = [
  {
      'name': 'apple',
      'quantity': 1
  },
  {
      'name': 'banana',
      'quantity': 1
  },
  {
      'name': 'pineapple',
      'quantity': 1
  },
]
for ingredient in ingredients_list:
  if not ingredient.get('quantity'):
    print(f'problem, no {ingredient.get("name")}')
    break
else:
  print('good to go')
```
