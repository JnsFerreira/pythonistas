```python
class Person:
  # Magic methods or Dunder methods
  def __init__(self, name: str, age: int = 15):
    self.name = name
    self.age = age
  
  def get_hobby(self):
    return 'play game'


class Student(Person):
  def __init__(self, *args, hobby: dict = None, **kwargs):
    # print(args) ('iago', 'alexandre', 'caio')
    # print(kwargs) {'name': 'jonas', 'age': '15'}
    self.hobby = hobby
    if not hobby:
      self.hobby = 'nao tenho hobby otario'
    
    super().__init__(*args, **kwargs)  # person.__init__()
  
  def get_hobby(self):
    return self.hobby


p = Person(name='veronica', age=18)
s = Student(name='jonas', age=21)


print(s.name, s.age, s.get_hobby())



# More dunder methods x
# Diff between class and static methods x
# Decorators

```
