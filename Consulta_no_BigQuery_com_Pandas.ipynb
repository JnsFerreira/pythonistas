````python
# Criar credencial de acesso
import google.oauth2.credentials

credentials = google.oauth2.credentials.Credentials('CREDENCIAL ADQUIRIDA PELO TERMINAL')
````

````python
# Instanciar cliente do BigQuery
from google.cloud import bigquery

client = bigquery.Client(credentials=credentials, project='white-script-238515')
````

````python
# Realizar a query.
query = '''
    SELECT * FROM `white-script-238515.clean_seller.seller_bank_account` 
    LIMIT 100
'''
query_job = client.query(query)  # API request
dataframe = query_job.to_dataframe()
````
