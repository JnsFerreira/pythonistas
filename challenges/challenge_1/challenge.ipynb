Desafio 1
 
A identificação dos usuários em um sistema é feita por uma string de tamanho variável entre 4 e 32.  
Exemplos: 1257
 
O último par é composto pelos digitos verificadores, o primeiro digito verificador é calculado da seguinte maneira:
 
- Separe os dígitos verificadores dos demais números:  
12 e 57  
 
- Cada um dos demais números deverão ser multiplicados por um peso e os resultados deverão ser somados.  
O peso vale 1 para o primeiro, 2 para o segundo e assim por diante:  
1\*1 + 2\*2 = 5
 
- A soma deverá ser dividida por 10 e o resto da divisão será o primeiro digito verificador:  
5/10, resto = 5
 
Primeiro digito verificador é **5**.
 
O segundo digito verificador é calculado de forma semelhante ao primeiro mas
o peso que cada numero deverá ser multiplicado se iniciará pelo valor do primeiro digito verificador:  
 
- 1\***5** + 2\***6** = 17
- 17/10, resto = 7
 
O segundo digito verificador é **7**
 
String de identificação **1257** é **válida**.
 
**OBS:** *As strings de verificação compostas por **todos os seus dígitos iguais** são consideradas **inválidas**.*
 
Dada essa regra de formação para a identificação dos usuários, crie a função chamada ***identification_validator*** que receba como parâmetro uma string de identificação e retorne um boolean indicando se é válida.
